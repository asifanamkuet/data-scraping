```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

def main():
    # Initialize ChromeDriver using WebDriverManager
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)

    # Open the Costco product page
    url = "https://www.costco.ca/.product.4000276522.html"
    driver.get(url)

    # Wait for the page to load
    time.sleep(3)

    # Accept cookies if necessary
    try:
        accept_cookies = driver.find_element(By.ID, "onetrust-accept-btn-handler")
        accept_cookies.click()
    except:
        pass  # Continue if the button is not found

    # Loop through all pages of reviews
    while True:
        print("enters loop")
        
        # Get the page source and parse it with BeautifulSoup
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')
        # print(soup)

        # Extract the product title
        product_title = soup.find("span", {"class": "product-title", "itemprop": "name"})
        if product_title:
            product_title = product_title.get_text(strip=True)
            print(f"Product Title: {product_title}")
        else:
            print("Product title not found")

        # Extract review titles and content
        reviews = soup.find_all("h3", {"class": "bv-content-title"})
        print(reviews)

        for review in reviews:
            print("enter reviews")
            review_title = review.find("h3", {"class": "bv-content-title"})
            review_content = review.find("div", {"class": "bv-content-summary-body"}).find("div", {"class": "bv-content-summary-body-text"}).find("p")
            print(review_title)
            print(review_content)
            
            if review_title:
                review_title = review_title.get_text(strip=True)
                print(f"Review Title: {review_title}")
            else:
                print("Review title not found")

            if review_content:
                review_content = review_content.get_text(strip=True)
                print(f"Review Content: {review_content}")
            else:
                print("Review content not found")
            print("-" * 80)

        # Check if there's a "Next" button to go to the next page
        try:
            next_button = driver.find_element(By.CLASS_NAME, "review-paginator-next")
            if "disabled" in next_button.get_attribute("class"):
                break  # Exit the loop if the next button is disabled (last page)
            next_button.click()
            time.sleep(3)  # Wait for the next page to load
        except:
            break  # Exit the loop if there's no next button

    # Close the WebDriver
    driver.quit()

if __name__ == "__main__":
    main()
```
